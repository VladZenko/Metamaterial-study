$\nabla \times \textbf{E} = -\mu \frac{\partial\textbf{H}}{\partial t} $

$\nabla \times \textbf{H} = \textbf{J} + \epsilon \frac{\partial\textbf{E}}{\partial t} $

*set $J$ to 0 for now(no currents)*

# RHS

Derivative approximations (at integer & half-integer steps)

$\frac{df}{dx}(x)=\frac{f\left(x+\frac{\Delta x}{2} \right) - f\left(x-\frac{\Delta x}{2} \right)}{\Delta x}$

$\frac{df}{dx}\left(x+\frac{\Delta x}{2} \right)=\frac{f\left(x+\Delta x\right) - f\left( x \right)}{\Delta x}$

Hence

$\frac{\partial \textbf{H}}{\partial t}(t) \approx \frac{\textbf{H}\left(t+\frac{\Delta t}{2} \right) - \textbf{H} \left(t-\frac{\Delta t}{2} \right)}{\Delta t} = -\frac{1}{\mu} \nabla \times \textbf{E}(t)$

Note that to solve, need to know H at half-integer time steps, and E at integer steps. to do that, define $\frac{\partial \textbf{E}}{\partial t}$ at $(t+\Delta t/2)$ rather than $t$

$\frac{\partial \textbf{E}}{\partial t}(t+\Delta t/2) \approx \frac{\textbf{E}\left(t+\Delta t \right) - \textbf{E} \left(t\right)}{\Delta t} = -\frac{1}{\epsilon} \nabla \times \textbf{H}(t+\Delta t/2)$

# LHS

$\nabla \times \textbf{E} = \begin{pmatrix} \frac{\partial E_z}{\partial y}-\frac{\partial E_y}{\partial z} \\ \frac{\partial E_x}{\partial z}-\frac{\partial E_z}{\partial x} \\ \frac{\partial E_y}{\partial x}-\frac{\partial E_x}{\partial y} \end{pmatrix}$

$\nabla \times \textbf{H} = \begin{pmatrix} \frac{\partial H_z}{\partial y}-\frac{\partial H_y}{\partial z} \\ \frac{\partial H_x}{\partial z}-\frac{\partial H_z}{\partial x} \\ \frac{\partial H_y}{\partial x}-\frac{\partial H_x}{\partial y} \end{pmatrix}$

Hence

$\frac{\partial H_x}{\partial t} = -\frac{1}{\mu} \left( \frac{\partial E_z}{\partial y}-\frac{\partial E_y}{\partial z} \right)$

$\frac{\partial H_y}{\partial t} = -\frac{1}{\mu} \left( \frac{\partial E_x}{\partial z}-\frac{\partial E_z}{\partial x} \right)$

$\frac{\partial H_z}{\partial t} = -\frac{1}{\mu} \left( \frac{\partial E_y}{\partial x}-\frac{\partial E_x}{\partial y} \right)$

And

$\frac{\partial E_x}{\partial t} = \frac{1}{\epsilon} \left( \frac{\partial H_z}{\partial y}-\frac{\partial H_y}{\partial z} \right)$

$\frac{\partial E_y}{\partial t} = \frac{1}{\epsilon} \left( \frac{\partial H_x}{\partial z}-\frac{\partial H_z}{\partial x} \right)$

$\frac{\partial E_z}{\partial t} = \frac{1}{\epsilon} \left( \frac{\partial H_y}{\partial x}-\frac{\partial H_x}{\partial y} \right)$

Hence, we need to solve 6 PDE's for 3D, 3 PDE's for 2D, 2 PDE's for 1D

# 1D

Assumptions:

$*)$ Propagating only in y-direction:

$\frac{\partial}{\partial x} = \frac{\partial}{\partial z} = 0$

$*)$ Field is x-polarised:

$E_y = E_z = 0$

Hence, the PDE's left:

$\frac{\partial H_z}{\partial t} = \frac{1}{\mu} \frac{\partial E_x}{\partial y}$

$\frac{\partial E_x}{\partial t} = \frac{1}{\epsilon} \frac{\partial H_z}{\partial y}$

Approximate the spatial derivatives

$\frac{\partial E_x}{\partial y} \approx \frac{E_x\left(y+\Delta y \right) - E_x \left(y\right)}{\Delta y} = \mu \frac{\partial H_z}{\partial t}(y+\Delta y/2)$

$\frac{\partial H_z}{\partial y} \approx \frac{H_z\left(y+\frac{\Delta y}{2} \right) - H_z \left(y-\frac{\Delta y}{2} \right)}{\Delta y} = \epsilon \frac{\partial E_x}{\partial t}(y)$

Need $H_z$ at half steps ; $E_x$ at integer steps

# Yee grid

$H_z$ defined at $t+\frac{\Delta t}{2}$ and $y+\frac{\Delta y}{2}$

$E_x$ defined at $t$ and $y$

Recall the PDE's we need to solve

$\text{1)\quad}   \frac{\partial H_z}{\partial t} = \frac{1}{\mu} \frac{\partial E_x}{\partial y}$

$\text{2)\quad} \frac{\partial E_x}{\partial t} = \frac{1}{\epsilon} \frac{\partial H_z}{\partial y}$

Redefine using approximation for derivatives

$\text{1)\quad} \mu \frac{H_z(y+\Delta y/2\ ;\ t+\Delta t/2)-H_z(y+\Delta y/2\ ;\ t-\Delta t/2)}{\Delta t} = \frac{E_x(y+\Delta y\ ;\ t)-E_x(y\ ;\ t)}{\Delta y}$

$\text{2)\quad} \epsilon \frac{E_x (y\ ;\ t+\Delta t) - E_x (y\ ;\ t)}{\Delta t} = \frac{H_z(y+\Delta y/2\ ;\ t+\Delta t/2)-H_z(y-\Delta y/2\ ;\ t+\Delta t/2)}{\Delta y}$

In computationa terms, let time run from $0$ to $n\Delta t$, and space from $0$ to $j\Delta y$

$\text{1)\quad} \mu \frac{H_z(j+1/2\ ;\ n+1/2)-H_z(j+1/2\ ;\ n-1/2)}{\Delta t} = \frac{E_x(j+1\ ;\ n)-E_x(j\ ;\ n)}{\Delta y}$

$\text{2)\quad} \epsilon \frac{E_x (j\ ;\ n+1) - E_x (j\ ;\ n)}{\Delta t} = \frac{H_z(j+1/2\ ;\ n+1/2)-H_z(j-1/2\ ;\ n+1/2)}{\Delta y}$

Python cannot interpret half integer indices (it simply makes no sense). Hence, from now on let $j+1/2 \rightarrow j$ and $n+1/2 \rightarrow n$

$\text{1)\quad} \mu \frac{H_z(j\ ;\ n)-H_z(j\ ;\ n-1)}{\Delta t} = \frac{E_x(j+1\ ;\ n)-E_x(j\ ;\ n)}{\Delta y}$

$\text{2)\quad} \epsilon \frac{E_x (j\ ;\ n+1) - E_x (j\ ;\ n)}{\Delta t} = \frac{H_z(j\ ;\ n)-H_z(j-1\ ;\ n)}{\Delta y}$

Rearrange to get fields for current indices, $(j, n)$ for $H_z$ and updated indices $(j, n+1)$ for $E_x$

$\text{1)\quad} H_z(j,n) = H_z(j,n-1) + \frac{\Delta t}{\mu \Delta y}(E_x(j+1,n)-E_x(j,n))$

$\text{2)\quad} E_x(j, n+1) = E_x(j,n) + \frac{\Delta t}{\epsilon \Delta y}(H_z(j,n) - H_z(j-1,n))$

Hence, for each step in time, update: 

$*)$ Magnetic field for $j = 0 \rightarrow j_{max}-1$

$*)$ Then, update using the electric field, for $j=1 \rightarrow j_{max}$

# 2D

$\nabla \times \textbf{E} = -\mu \frac{\partial\textbf{H}}{\partial t} $

$\nabla \times \textbf{E} = \begin{pmatrix} \frac{\partial E_z}{\partial y}-\frac{\partial E_y}{\partial z} \\ \frac{\partial E_x}{\partial z}-\frac{\partial E_z}{\partial x} \\ \frac{\partial E_y}{\partial x}-\frac{\partial E_x}{\partial y} \end{pmatrix}$

$\frac{\partial E_x}{\partial t} = \frac{1}{\epsilon} \left( \frac{\partial H_z}{\partial y}-\frac{\partial H_y}{\partial z} \right)$

$\frac{\partial E_y}{\partial t} = \frac{1}{\epsilon} \left( \frac{\partial H_x}{\partial z}-\frac{\partial H_z}{\partial x} \right)$

$\frac{\partial E_z}{\partial t} = \frac{1}{\epsilon} \left( \frac{\partial H_y}{\partial x}-\frac{\partial H_x}{\partial y} \right)$

$\nabla \times \textbf{H} = \textbf{J} + \epsilon \frac{\partial\textbf{E}}{\partial t} $

$\nabla \times \textbf{H} = \begin{pmatrix} \frac{\partial H_z}{\partial y}-\frac{\partial H_y}{\partial z} \\ \frac{\partial H_x}{\partial z}-\frac{\partial H_z}{\partial x} \\ \frac{\partial H_y}{\partial x}-\frac{\partial H_x}{\partial y} \end{pmatrix}$

$\frac{\partial H_x}{\partial t} = -\frac{1}{\mu} \left( \frac{\partial E_z}{\partial y}-\frac{\partial E_y}{\partial z} \right)$

$\frac{\partial H_y}{\partial t} = -\frac{1}{\mu} \left( \frac{\partial E_x}{\partial z}-\frac{\partial E_z}{\partial x} \right)$

$\frac{\partial H_z}{\partial t} = -\frac{1}{\mu} \left( \frac{\partial E_y}{\partial x}-\frac{\partial E_x}{\partial y} \right)$

Assumptions:

$\text{*)\quad} \frac{\partial}{\partial z} = 0$ 

$\text{*)\quad} E_z = 0$


Hence, PDE's left to solve are:

$\text{1)\quad} \frac{\partial E_x}{\partial t} = \frac{1}{\epsilon}  \frac{\partial H_z}{\partial y}$

$\text{2)\quad} \frac{\partial E_y}{\partial t} = -\frac{1}{\epsilon} \frac{\partial H_z}{\partial x}$

$\text{3)\quad} \frac{\partial H_z}{\partial t} = -\frac{1}{\mu} \left( \frac{\partial E_y}{\partial x}-\frac{\partial E_x}{\partial y} \right)$

$\text{1)\quad} \epsilon \frac{E_x (x\ ;\ y\ ;\ t+\Delta t) - E_x (x\ ;\ y\ ;\ t)}{\Delta t} = \frac{H_z(x+\Delta x/2\ ;\ y+\Delta y/2\ ;\ t+\Delta t/2)-H_z(x+\Delta x/2\ ;\ y-\Delta y/2\ ;\ t+\Delta t/2)}{\Delta y}$

$\text{2)\quad} \epsilon \frac{E_y (x\ ;\ y\ ;\ t+\Delta t) - E_y (x\ ;\ y\ ;\ t)}{\Delta t} = -\frac{H_z(x+\Delta x/2\ ;\ y+\Delta y/2\ ;\ t+\Delta t/2)-H_z(x-\Delta x/2\ ;\ y+\Delta y/2\ ;\ t+\Delta t/2)}{\Delta x}$

$\text{3)\quad} \mu \frac{H_z(x+\Delta x/2\ ;\ y+\Delta y/2\ ;\ t+\Delta t/2)-H_z(x+\Delta x/2\ ;\ y+\Delta y/2\ ;\ t-\Delta t/2)}{\Delta t} = \frac{E_x(x\ ;\ y+\Delta y\ ;\ t)-E_x(x\ ;\ y\ ;\ t)}{\Delta y} - \frac{E_y(x+\Delta x\ ;\ y\ ;\ t)-E_y(x\ ;\ y \ ;\ t)}{\Delta x}$

In terms of computational indices:

$x, y, t \rightarrow i, j, n$

$\Delta (x,y,n)/2 \rightarrow \frac{1}{2}$

$\Delta (x,y,n) \rightarrow 1$

$\text{1)\quad} \epsilon \frac{E_x (i\ ;\ j\ ;\ n+1) - E_x (i\ ;\ j\ ;\ n)}{\Delta t} = \frac{H_z(i+1/2\ ;\ j+1/2\ ;\ n+1/2)-H_z(i+1/2\ ;\ j-1/2\ ;\ n+1/2)}{\Delta y}$

$\text{2)\quad} \epsilon \frac{E_y (i\ ;\ j\ ;\ n+1) - E_y (i\ ;\ j\ ;\ n)}{\Delta t} = -\frac{H_z(i+1/2\ ;\ j+1/2\ ;\ n+1/2)-H_z(i-1/2\ ;\ j+1/2\ ;\ n+1/2)}{\Delta x}$

$\text{3)\quad} \mu \frac{H_z(i+1/2\ ;\ j+1/2\ ;\ n+1/2)-H_z(i+1/2\ ;\ j+1/2\ ;\ n-1/2)}{\Delta t} = \frac{E_x(i\ ;\ j+1\ ;\ n)-E_x(i\ ;\ j\ ;\ n)}{\Delta y} - \frac{E_y(i+1\ ;\ j\ ;\ n)-E_y(i\ ;\ j \ ;\ n)}{\Delta x}$

Write in terms of whole indices for computational purposes:

$i+1/2 \rightarrow i \ ;\  j+1/2 \rightarrow j \ ;\  n+1/2 \rightarrow n$

$\text{1)\quad} \epsilon \frac{E_x (i\ ;\ j\ ;\ n+1) - E_x (i\ ;\ j\ ;\ n)}{\Delta t} = \frac{H_z(i\ ;\ j\ ;\ n)-H_z(i\ ;\ j-1\ ;\ n)}{\Delta y}$

$\text{2)\quad} \epsilon \frac{E_y (i\ ;\ j\ ;\ n+1) - E_y (i\ ;\ j\ ;\ n)}{\Delta t} = -\frac{H_z(i\ ;\ j\ ;\ n)-H_z(i-1\ ;\ j\ ;\ n)}{\Delta x}$

$\text{3)\quad} \mu \frac{H_z(i\ ;\ j\ ;\ n)-H_z(i\ ;\ j\ ;\ n-1)}{\Delta t} = \frac{E_x(i\ ;\ j+1\ ;\ n)-E_x(i\ ;\ j\ ;\ n)}{\Delta y} - \frac{E_y(i+1\ ;\ j\ ;\ n)-E_y(i\ ;\ j \ ;\ n)}{\Delta x}$

Rearrange to get fields for current indices, $(i, j, n)$ for $H_z$ and updated indices $(i, j, n+1)$ for $E_x$ & $E_y$

$\text{1)\quad}  E_x(i\ ;\ j\ ;\ n+1) = E_x(i\ ;\ j\ ;\ n) + \frac{\Delta t}{\epsilon \Delta y} \left[ H_z(i\ ;\ j\ ;\ n) - H_z(i\ ;\ j-1\ ;\ n) \right]$

$\text{2)\quad}  E_y(i\ ;\ j\ ;\ n+1) = E_y(i\ ;\ j\ ;\ n) - \frac{\Delta t}{\epsilon \Delta x} \left[ H_z(i\ ;\ j\ ;\ n) - H_z(i-1\ ;\ j\ ;\ n) \right]$
