```python
import sympy
from sympy.vector import *
from IPython.display import Math
sympy.init_printing(use_unicode=True)
```

```python
U1, U2, u, u_pr, xs, ys = sympy.symbols('U1, U2, u, u^{\prime}, x, y')
```

```python
import numpy as np
from numpy import *
```

```python
grid = np.linspace(-1,1,400)
x,y = np.meshgrid(grid, grid)
```

```python
U1_val = 1.1
U2_val = 1.2

u_pr = ((u-U1)*U2)/(U2-U1)
```

```python
E00 = 1/u * (U2-U1)/(U2)
E11 = U2/(U2-U1) * 1/(u_pr*u)
E22 = U2/(U2-U1) *u_pr/u
```

```python
E00 = E00.subs(u, sympy.sqrt(xs**2+ys**2))
E11 = E11.subs(u, sympy.sqrt(xs**2+ys**2))
E22 = E22.subs(u, sympy.sqrt(xs**2+ys**2))
```

```python
E00 = E00.subs(xs,sympy.symbols('(x)'))
E00 = E00.subs(ys,sympy.symbols('(y)'))
E00 = E00.subs(U1,sympy.symbols('(U1_val)'))
E00 = E00.subs(U2,sympy.symbols('(U2_val)'))

E11 = E11.subs(xs,sympy.symbols('(x)'))
E11 = E11.subs(ys,sympy.symbols('(y)'))
E11 = E11.subs(U1,sympy.symbols('(U1_val)'))
E11 = E11.subs(U2,sympy.symbols('(U2_val)'))

E22 = E22.subs(xs,sympy.symbols('(x)'))
E22 = E22.subs(ys,sympy.symbols('(y)'))
E22 = E22.subs(U1,sympy.symbols('(U1_val)'))
E22 = E22.subs(U2,sympy.symbols('(U2_val)'))
```

```python
E22 = eval(str(E22))
E11 = eval(str(E11))
E00 = eval(str(E00))
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (12,16))

ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

ax1.imshow(E00,cmap='jet')
ax1.set_title(r'$\epsilon^{00}$')
ax2.imshow(E11,cmap='jet')
ax2.set_title(r'$\epsilon^{11}$')
ax3.imshow(E22,cmap='jet')
ax3.set_title(r'$\epsilon^{22}$')


plt.tight_layout()
```
