```python
import numpy as np
g = np.linspace(-10,10,400)
xg, yg = np.meshgrid(g, g)
zg = 0
```

```python
M = 1
```

```python
E00 = ((-M+(xg**2+yg**2+zg**2)**0.5)**1.5)/((xg**2+yg**2)**0.25*(xg**2+yg**2+zg**2)**1.5)
```

```python
import matplotlib.pyplot as plt

plt.imshow(E00, cmap='jet')
```

```python
import sympy as sp
```

```python
xs, ys, zs, Ms = sp.symbols('x y z M')
```

```python
E00_sympy = ((-Ms+(xs**2+ys**2+zs**2)**0.5)**(3/2))/((xs**2+ys**2)**0.25*(xs**2+ys**2+zs**2)**1.5)
E00_sympy
```

```python
E00_sympy = E00_sympy.subs(Ms,sp.symbols('M'))
E00_sympy = E00_sympy.subs(xs,sp.symbols('(xg)'))
E00_sympy = E00_sympy.subs(ys,sp.symbols('(yg)'))
E00_sympy = E00_sympy.subs(zs,sp.symbols('(zg)'))

Intensity_E00 = eval(str(E00_sympy))
```

```python
plt.imshow(Intensity_E00, cmap='jet')
```
